## Practice on Beautifulsoup

```python
from bs4 import BeautifulSoup
import requests

url="https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YBkllW5uJPY"

raw_data=requests.get(url)
#print(raw_data.text)

soup= BeautifulSoup(raw_data.text, "html.parser")
#print(soup)

print(soup.prettify())
```

```python
day_7=soup.find_all("div", class_="tombstone-container")
print(day_7)
```

```python
for i in (range(len(day_7))):
    print('#########')
    print(day_7[i])
```

```python
## Lets see the inormation about tonight

tonight=day_7[0]
print(tonight)  # We used prettify before
```

```python
period=tonight.find(class_="period-name")
print(period)
print(period.text)
print(period.get_text())
```

```python
short_desc=tonight.find(class_="short-desc")
print(short_desc)
```
